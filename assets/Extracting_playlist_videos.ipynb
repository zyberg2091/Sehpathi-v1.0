```
import csv 
import json



def data_prep(list_data,f_name,playlist_ids):
    for course in list_data:
      with open(course,'r',encoding='utf-8') as output_file:
          files=json.load(output_file)
      items=files['items']
      for item in items:
        playlist_ids.append(item['id']['playlistId'])

    return playlist_ids

if __name__=="__main__":
  sql_playlist_ids,py_playlist_ids,js_playlist_ids=[],[],[]
  for i in range(3):
    f_name=str(input())
    if f_name=='js':
      js_playlist_ids = data_prep(['javascript1.json','javascript2.json','javascript3.json','javascript4.json','javascript5.json'],f_name,js_playlist_ids)
    elif f_name=='python':
      py_playlist_ids = data_prep(['pycourse1.json','pycourse2.json','pycourse3.json','pycourse4.json','pycourse5.json'],f_name,py_playlist_ids)
    elif f_name=='sql':
      sql_playlist_ids = data_prep(['sql1.json','sql2.json','sql3.json','sql4.json','sql5.json'],f_name,sql_playlist_ids)
  
```
